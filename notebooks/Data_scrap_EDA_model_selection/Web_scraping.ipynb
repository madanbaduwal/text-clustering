# Load html raw data

I have done secure copy(scp) of doubt data(jun,july,aug)from aws instance.

```python
import os 
import glob

# We are currefntly at notebook folder in project structure. We need to move toward data/raw folder 
# This is also one time run code , if you run it again it doesn't find file because it already moved two step back , so need restart
print(os.chdir("../../data/raw"))
print(os.getcwd())

raw_data_path = os.getcwd()
month_folders = os.listdir(os.getcwd())
all_folders_files = []
for month_folder in month_folders:
    files = glob.glob(raw_data_path +f'//{month_folder}' + '/*.html') # in window we need // in other os ot may be different 
    all_folders_files +=  files

# At this time we get the location of all files
```

**Non-necessary informations in html :**

```python
import re 
import urllib.request 
from bs4 import BeautifulSoup 

def scrap_without_tags(url):
    #html = urllib.request.urlopen(url)  # if you are trying to extract from url
    
    # Below two line code are for open local file and store in variable
    # exact location need to be mentioned
    HTMLFileToBeOpened = open(url,encoding="utf8")
    # Reading the file and storing in a variable
    contents = HTMLFileToBeOpened.read()
    
    soup = BeautifulSoup(contents,'html.parser')
#     data = soup.get_text("|",strip=True) 
#     data = data.split("|")    
    return soup
```

```python
soup = scrap_without_tags(all_folders_files[1])
```

```python
all_folders_files
```

```python
# Extract all tag in the html
for tag in soup.find_all(True):
    print(tag.name)
```

```python
# extract url from file name
all_folders_urls = []
for file in all_folders_files:
    url = file.split('\\')[-1].split('.')[0]
    url = url.replace("_",".") # I found that the name of file is the url name with . is replace by _, so replace _ by .
    all_folders_urls.append(url)

# This is for 
contents = []
for url in all_folders_files:
    content = scrap_without_tags(url)
    contents.append(content)
```

**Save scraped data**

```python
import pandas as pd 

print(os.chdir("../interim"))
print(os.getcwd())

df = pd.DataFrame({'url': all_folders_urls,'doubt_withought_tags':contents}) 
df.to_csv('doubt_withought_tags.csv', index=False, encoding='utf-8')
data = pd.read_csv("./doubt_withought_tags.csv")
data.head(30)
```
